# Bioferineries

The `biorefineries` package comes with key BioSTEAM biorefinery designs, including a design for 2nd generation ethanol production from corntover and a design for the co-production of ethanol and biodiesel from lipidcane [[1, 2]](#References). To load the cornstover biorefinery, simply load the `system`:

```python
import biosteam as bst
# First "pip install biorefineries"
from biorefineries.cornstover import system
```

Now your flowsheet should change to the cornstover biorefinery:

```python
bst.find
```

All systems, unit operations, and streams are registered in your flowsheet:

```python
bst.find.system
```

By convention, the complete biorefinery system begins with the same name as the flowsheet:

```python
cornstover_sys = bst.find.system.cornstover_sys
cornstover_sys.diagram()
```

Similarly, you can also load the lipidcane biorefinery:

```python
from biorefineries.lipidcane import system
cornstover_sys = bst.find.system.lipidcane_sys
cornstover_sys.diagram()
```

Note that your flowsheet is now "lipidcane":

```python
bst.find
```

You can switch flowsheets anytime with the `set_flowsheet` method:

```python
bst.find.flowsheet
```

```python
bst.find.set_flowsheet(bst.find.flowsheet.default)
```

```python
bst.find
```

You can also import only the Species object used to model the thermodynamic properties, without having to load the whole biorefinery system:

```python
from biorefineries.lipidcane import species 
# Note: The species module contains all species objects
species.lipidcane_species
```
