# Importing biorefineries

The `biorefineries` package includes a design for 2nd generation ethanol production from corntover and a design for the co-production of ethanol and biodiesel from lipidcane [[1, 2]](#References). To load the cornstover biorefinery, simply import the system:

```python
import biosteam as bst
from biorefineries import cornstover as cs
cs.load() # Load biorefinery
```

Now your flowsheet should change to the cornstover biorefinery:

```python
bst.main_flowsheet
```

All systems, unit operations, and streams are registered in your flowsheet:

```python
bst.main_flowsheet.system
```

By convention, the complete biorefinery system begins with the same name as the flowsheet:

```python
cs.cornstover_sys.diagram() # Imported in line 1
```

Similarly, you can also load the lipidcane biorefinery:

```python
from biorefineries import lipidcane as lc
lc.load() # Load biorefinery
lc.lipidcane_sys.diagram()
```

Note that your main flowsheet is now "lipidcane":

```python
bst.main_flowsheet
```

You can switch flowsheets anytime with the `set_flowsheet` method:

```python
bst.main_flowsheet.flowsheet
```

```python
bst.main_flowsheet.set_flowsheet('default')
```

```python
bst.main_flowsheet
```

You can also import only the Chemicals object used to model the thermodynamic properties, without having to load the whole biorefinery system:

```python
from biorefineries.lipidcane import chemicals
chemicals
```
